### Objects
- Everything in Python is a object
- Python compiles the Literal - (10) and creates a (Integer) Object that has 10 as it's state
- Entity: Anything that exists in code (Class, object, variable, literal, etc.)
- Class: A class is a blueprint for creating objects. it defines State: The data attributes that each object will hold. Methods: The behaviors (functions) that operate on that data.
- Object: An entity created from a class, with state and methods
- In Python, objects can already exist or can be created
#### Objects That Already Exist
  - Literals like 10, 3.14, "hello", True, None
  - Built-in types like int, str, list, dict, bool
  - Functions like print(), len(), type()
  - Classes like object, Exception, int, str
#### Objects You Create
```python
class Person:
    def __init__(self, name):
        self.name = name
p = Person("Alice")  # Creates a new Person object
```
- Python `(objects)` have `State(data/literals) and methods(functionality-can have parameters) are Atributes` called Encapsulation 
> Example: `object with type Integer` have `Value` and `.__add__(Parameter) `
- `Encapsulation` means grouping related data and method together inside a class, and `controlling access` to that data.
- Dot notation to access attributes of an object
- Mutable Data Types(attributes/internal state can be changed) and Immutable Data Types(internal state cannot be changed)
-  State: Refers to the internal data stored in an object. It can change over time as methods are called
-  Method: is a function that can access and modify the object’s state.
> Example: `(10).__add__(100) = 110`
> 
> `(state).__Method__(Parameter)`

### Variables
- Naming objects/labels/symbols
- `=` Assignment Operator, assigns/references/points the object to a label/variable
> balance = 1_000.00
- Python compiles the Literal - (10) and creates a (Integer) Object in the memory that has 10 as it's state
> x = 10
-  when we label that object as `x`, then x refrences  that object 10
-  python evaluates RHS first then assigns the result to LHS.
> x = 'shiva'
- References can change over time, we can point the x to 'shiva' now.
- Objects like `lists` are mutable(object's state can be appended/changed), but the reference didn't change here.
> x = float(10)
- This will create a float object
  
#### Variable Names
- case sensitive
- Must start with a letter (a–z, A–Z) or underscore _ or unicode characters
- Can contain letters, digits, and underscores
- Cannot start with a digit
- Cannot use Python keywords (like class, def, if, else, etc.)
- Use descriptive names: total_price, user_name
- Use snake_case for variables: user_age, is_valid for variables and CamelCase: AccountBalance, accountBalance for Classes
- Constants are usually written in UPPER_CASE: PI = 3.14
- Avoid single-letter names unless in loops: i, j
- SyntaxError: Invalid Syntax
